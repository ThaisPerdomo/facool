# Filtrando resultados

```python
import pandas as pd
dados = pd.read_csv('aluguel.csv', sep = ';')
```

```python
dados.head(10) # Checando as 10 primeiras linhas do DataFrame, pra ver se está tudo certo
```

```python
list(dados['Tipo'].drop_duplicates()) 
# Lista de tipos de imóveis, excluindo os duplicados
```

## Exemplo 1

```python
# Criando uma lista de tipos de imóveis residenciais, para filtrar o DataFrame

residencial = ['Quitinete', 
'Casa',
'Apartamento',
'Casa de Condomínio',
'Casa de Vila']
```

```python
dados['Tipo'].isin(residencial).head(10)

# Procurando nos 10 primeiros registros quais são residenciais
```

```python
selecao = dados['Tipo'].isin(residencial)

# Atribuindo a seleção a uma variável. Agora sim, vamos filtrar o DataFrame
```

```python
dados_residencial = dados[selecao]

dados_residencial
```

## Exemplo 2

```python
casas = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
selecao_casas = dados['Tipo'].isin(casas)

selecao_casas
```

```python

show_all_houses = dados[selecao_casas]

show_all_houses
```

# Exportando a base de dados filtrada

```python
show_all_houses.to_csv('casas_para_alugar.csv', sep = ';', index = False)
```
